# NRES Tess input catalog worksheets

Small routinex to access the TESS input catalog. Working of a subsample of the full catlog, selected to have Tmag 
< 10 or so. Loaded in from very telling file named test.dat

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt ('test.dat', delimiter=',',dtype=None, skip_header=0, names=True)
```

```python
plt.plot (data['Teff'],data['logg'], '.')
```

```python
teffbins=[0, 4000,4370,4790,5250,5750, 6300, 7580, 8320, 20000]
```

```python
loggcond = [ (data['logg']>3.5 ), (data['logg']<2) & (data['logg'] > 0.5) ] 
teffcond = list( ( (data['Teff'] > teffbins[x]) & (data['Teff']<=teffbins[x+1])) for x in range (0,len(teffbins)-1))

```

```python
for logg in loggcond:
    for teff in teffcond:
      
        cond = logg & teff
        plt.plot (data['ra'][cond], data['dec'][cond], ".")
```

```python
for teff in teffcond:
    cond = loggcond[1] & teff & ( (data['dec']) < 10) & (data['dec'] >-50)
    plt.plot (data['ra'][cond], data['dec'][cond], ".")
    
    for (ra,dec,Tmag,teff,logg) in zip(data['ra'][cond], data['dec'][cond], data['Tmag'][cond], data['Teff'][cond], data['logg'][cond]):
        print ra,dec,Tmag,teff,logg
    
```
